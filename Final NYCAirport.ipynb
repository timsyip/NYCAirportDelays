# Cancelled

## All Airports

```python
#All Flights Cancelled in Dataframe
pct_cancelled= NYC['Cancelled'].mean()*100
pct_cancelled
```

```python
# Number of cancelled flights from all years by airline
Cancelled_count=NYC.query('Cancelled')['Airline'].value_counts()
Cancelled_count
```

```python
#Cancelled by Year
Cancelled_countyear = NYC[NYC['Cancelled'] == 1].groupby([NYC['FlightDate'].dt.year, 'Airline'])['Airline'].count().groupby(level=0).nlargest(10)
Cancelled_countyear
```

```python
# Cancelled Flights by Airlines and Percentage sorted by the most flights
cancelled=NYC.groupby(['Airline', 'Cancelled']).size().unstack().reset_index()
cancelled = cancelled.rename(columns={False: 'Not_Cancelled', True: 'Cancelled'})
cancelled['Total'] = cancelled['Not_Cancelled'] + cancelled['Cancelled']
cancelled['Percent_Cancelled'] = cancelled['Cancelled'] / cancelled['Total'] * 100
cancelled=cancelled.sort_values('Total', ascending=False)
cancelled
```

```python
#Most Cancelled by Years
for year in NYC['FlightDate'].dt.year.unique():
    cancelledyear = NYC[(NYC['Cancelled'] == True) & (NYC['FlightDate'].dt.year == year)].groupby(['Airline', 'Origin', 'Dest']).size().reset_index(name='Count').sort_values('Count', ascending=False).head(10)
    cancelledyear['Rank'] = cancelledyear['Count'].rank(method='dense', ascending=False)
    print("Top 10 Cancelled Flights for {}: \n{}".format(year, cancelledyear))
```

```python
# Number of Cancelled flight by Flight numbers
cancelledfltnum=NYC[NYC['Cancelled'] == True].groupby(['Airline', 'FlightNumber', 'Origin', 'Dest']).size().reset_index(name='Count').sort_values('Count', ascending=False).head(11)
cancelledfltnum['Rank'] = cancelledfltnum['Count'].rank(method='dense',ascending=False)
cancelledfltnum
```

```python
#Top 10 Cancelled Flights Numbers for each year

for year in NYC['FlightDate'].dt.year.unique():
    cancelledyearflightnum = NYC[(NYC['Cancelled'] == True) & (NYC['FlightDate'].dt.year == year)].groupby(['Airline', 'Origin', 'Dest', 'FlightNumber']).size().reset_index(name='Count').sort_values('Count', ascending=False).head(25)
    cancelledyearflightnum['Rank'] = cancelledyearflightnum['Count'].rank(method='dense', ascending=False)

    print("Top 10 Cancelled Flights Numbers for {}: \n{}".format(year, cancelledyearflightnum))
```
